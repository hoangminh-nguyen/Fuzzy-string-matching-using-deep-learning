# Plot results

This notebook plots the results of model inference done in the `Fig2_EMNLP_inference` notebook. The figures are the same as Fig. 2 in the following paper:

```
Hosseini, Nanni and Coll Ardanuy (2020), DeezyMatch: A Flexible Deep Learning Approach to Fuzzy String Matching, EMNLP: System Demonstrations.
```

---

In this notebook:

* skyline1: trained on *OCR* dataset
* skyline2: trained on *WG:en+OCR* dataset
* baseline: trained on *WG:en* dataset

---

* model A: both embedding and recurrent units are frozen (i.e., their parameters are not updated during fine-tuning).
* model B: only the embedding layer is frozen. 

---

To show the impact of fine-tuning and choice of architecture on the model performance, we trained various models starting with the baseline model and included more training instances from the training set of *OCR*.

The performance of these models is then assessed on the *OCR* test set. 

Refer to the paper for more information.

````python
plt.figure(figsize=(12, 10))

plt.plot(FT_results_lstm_v001[:, 0], FT_results_lstm_v001[:, 1], 
         '-o', color='tab:blue', lw=6, markersize=10, 
         label="LSTM", zorder=10)

plt.plot(FT_results_gru_v001[:, 0], FT_results_gru_v001[:, 1], 
         '-o', color='tab:red', lw=6, markersize=10, 
         label="GRU", zorder=10)

plt.plot(FT_results_rnn_v001[:, 0], FT_results_rnn_v001[:, 1], 
         '-o', color='tab:green', lw=6, markersize=10, 
         label="RNN", zorder=10)

plt.axhline(skyline1, 0, 1, 
            ls='--', color='k', lw=4,
            label="skyline 1")
plt.axhline(skyline2, 0, 1, 
            ls='-.', color='k', lw=4, 
            label="skyline 2")
plt.axhline(baseline1_gru, 0, 1, 
            ls=':', color='k', lw=4,
            label="baseline")

plt.xlabel("#entries", size=30)
plt.ylabel("F1 score", size=30)
plt.ylim(0.44)
plt.xticks([250, 8000, 16000, 32000, 64000, 84000], size=24, rotation=90)
plt.yticks(size=24)

plt.title("model A", size=32)

plt.grid()
plt.legend(fontsize=24)
plt.show()
````
